```python
temp_mean = acc_df['Temperature(F)'].mean()
acc_df['Temperature(F)'].fillna(temp_mean, inplace = True)
```

```python
wind_mean = acc_df['Wind_Chill(F)'].mean()
acc_df['Wind_Chill(F)'].fillna(wind_mean, inplace = True)

hum_mean = acc_df['Humidity(%)'].mean()
acc_df['Humidity(%)'].fillna(hum_mean, inplace = True)

pr_mean = acc_df['Pressure(in)'].mean()
acc_df['Pressure(in)'].fillna(pr_mean, inplace = True)

vis_mean = acc_df['Visibility(mi)'].mean()
acc_df['Visibility(mi)'].fillna(vis_mean, inplace = True)

wsp_mean = acc_df['Wind_Speed(mph)'].mean()
acc_df['Wind_Speed(mph)'].fillna(wsp_mean, inplace = True)

pt_mean = acc_df['Precipitation(in)'].mean()
acc_df['Precipitation(in)'].fillna(pt_mean, inplace = True)
```
