```python
# Grouping the DataFrame by Region of incident:

df.groupby('Region of Incident')
```

```python
df.groupby(['Incident year']).sum()['Total Number of Dead and Missing'].sort_values(ascending  = False)
```

```python
year_group = df.groupby(['Incident year']).sum()['Total Number of Dead and Missing'].sort_values(ascending  = False)
```

```python
year_group.plot(kind = 'bar', xlabel = 'Year of Incident', ylabel = 'Total Number of Dead and Missing', title = 'Total Number of Dead and Missing by Year')
```

```python
df.groupby(['Reported Month']).sum()['Total Number of Dead and Missing'].sort_values(ascending  = False)
```

```python
month_grp = df.groupby(['Reported Month']).sum()['Total Number of Dead and Missing'].sort_values(ascending  = False)
```

```python
month_grp.plot(kind = 'bar', xlabel = 'Reported Month', ylabel = 'Total Number of Dead and Missing', title = 'Total Number of Missing and Dead by Month')
```
